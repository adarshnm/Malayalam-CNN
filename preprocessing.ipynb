```python
import numpy as np
import os
import random
from tqdm import tqdm
import cv2
import pickle
```

```python
DATA_DIR='data/train_data/'
DIR=''
IMG_SIZE=86
LR=1e-3
MODEL_NAME='MHCR.{}-{}-model'.format(LR,'2dconv')
```

```python
def create_all_data():
    dataset = []
    op_class_num=0
    k=0
    DATA_FOLDER = os.listdir(DATA_DIR)
    for subdir in tqdm(DATA_FOLDER):
        label=[0 for i in range(44)]
        #print(subdir)
        index=int(subdir.split('R')[1])
        label[index-1]=1
        #print('label of ',subdir, label)
        #if k<4:
        dir_path=os.path.join(DATA_DIR,subdir)
        for img in os.listdir(dir_path):
                #print(img)
                img_path=os.path.join(dir_path,img)
                image=cv2.imread(img_path,0)
                dataset.append([np.array(image),np.array(label)])
        k+=1
    return dataset

    #print(dataset)
        
```

```python
dataset = create_all_data()
print(len(dataset))
```

```python
random.shuffle(dataset)
```

```python
picklefile = open("dataset.pickle", "wb")
pickle.dump(dataset,picklefile)
picklefile.close()
```
